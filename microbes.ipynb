## **Microbes in Human Welfare**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
import warnings
warnings.filterwarnings('ignore')

```

## 1. Microbial Population Growth (Exponential Growth)
This section analyzes the exponential growth of microbial populations over time using a mathematical model. The dataset contains simulated data on microbial growth in a controlled environment.



```python
# Exponential growth analysis
time = np.linspace(0, 24, 100)  # Time in hours
N0 = 100  # Initial population
r = 0.5  # Growth rate


population = N0 * np.exp(r * time)

df_growth = pd.DataFrame({'Time (hours)': time, 'Population Size': population})


plt.figure(figsize=(10, 6))
sns.lineplot(x='Time (hours)', y='Population Size', data=df_growth)
plt.title('Exponential Growth of Microbial Population')
plt.xlabel('Time (hours)')
plt.ylabel('Population Size')
plt.show()
```

## 2. Logistic Growth Curve
Logistic growth is another common model where microbial populations grow exponentially at first, then slow as they approach the carrying capacity of their environment.



```python
# Logistic growth analysis
K = 1000  # Carrying capacity(code by Anshuman Sinha)
population_logistic = K / (1 + ((K - N0)/N0) * np.exp(-r * time))


df_growth['Logistic Population'] = population_logistic

# Plot logistic growth
plt.figure(figsize=(10, 6))
sns.lineplot(x='Time (hours)', y='Logistic Population', data=df_growth, label='Logistic Growth')
plt.title('Logistic Growth of Microbial Population')
plt.xlabel('Time (hours)')
plt.ylabel('Population Size')
plt.show()
```

## 3. Fermentation Rate Using Michaelis-Menten Kinetics
This section uses the Michaelis-Menten kinetics model to analyze the fermentation rate of microbes in response to varying substrate concentrations.



```python
# Michaelis-Menten kinetics
S = np.linspace(0, 100, 100)  
Vmax = 1.5 
Km = 30  

# Michaelis-Menten equation
V = (Vmax * S) / (Km + S)

df_fermentation = pd.DataFrame({'Substrate Concentration': S, 'Fermentation Rate': V})

plt.figure(figsize=(10, 6))
sns.lineplot(x='Substrate Concentration', y='Fermentation Rate', data=df_fermentation)
plt.title('Fermentation Rate Using Michaelis-Menten Kinetics')
plt.xlabel('Substrate Concentration')
plt.ylabel('Fermentation Rate')
plt.show()
```

## 4. Clustering of Microbial Strains (K-Means)
Using K-means clustering, we can group similar microbial strains based on their biochemical properties (e.g., growth rate, enzyme production). This helps in categorizing microbes into different classes based on similarity.



```python
data = np.random.rand(100, 2)  # Random data representing microbial features

# K-Means clustering(code by Anshuman Sinha)
kmeans = KMeans(n_clusters=3, random_state=0).fit(data)
labels = kmeans.labels_

df_clusters = pd.DataFrame(data, columns=['Growth Rate', 'Enzyme Production'])
df_clusters['Cluster'] = labels

# Visualize the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Growth Rate', y='Enzyme Production', hue='Cluster', data=df_clusters, palette='Set1')
plt.title('K-Means Clustering of Microbial Strains')
plt.xlabel('Growth Rate')
plt.ylabel('Enzyme Production')
plt.show()
```

## Conclusion:
1. Microbial Population Growth (Exponential Growth): The exponential growth graph illustrates how microbial populations increase rapidly over time under optimal conditions. This model is essential for understanding the potential rapid proliferation of microbes when resources are abundant.

2. Logistic Growth Curve: The logistic growth curve shows how microbial populations grow quickly at first but slow down as they approach the carrying capacity of their environment. This model provides insights into how resource limitations affect growth and stabilization.

3. Fermentation Rate Using Michaelis-Menten Kinetics: The Michaelis-Menten kinetics graph reveals how the fermentation rate varies with substrate concentration. It highlights the enzyme efficiency and the saturation point where additional substrate does not significantly increase the rate.

4. Clustering of Microbial Strains (K-Means): The K-means clustering analysis groups microbial strains into distinct clusters based on their biochemical properties. This method helps in identifying and categorizing strains with similar characteristics for further study or practical applications.
